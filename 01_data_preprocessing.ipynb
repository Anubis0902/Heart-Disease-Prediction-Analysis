# Heart Disease Prediction System - Data Preprocessing
This notebook covers the data preprocessing steps for the Heart Disease dataset.


## Feature Explanations
- **age**: Age of the patient in years
- **sex**: Gender of the patient (1 = male, 0 = female)
- **cp**: Chest pain type (0 = typical angina, 1 = atypical angina, 2 = non-anginal pain, 3 = asymptomatic)
- **trestbps**: Resting blood pressure (in mm Hg on admission to the hospital)
- **chol**: Serum cholesterol in mg/dl
- **fbs**: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
- **restecg**: Resting electrocardiographic results (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy)
- **thalach**: Maximum heart rate achieved during exercise
- **exang**: Exercise induced angina (1 = yes; 0 = no)
- **oldpeak**: ST depression induced by exercise relative to rest
- **slope**: The slope of the peak exercise ST segment (0 = upsloping, 1 = flat, 2 = downsloping)
- **ca**: Number of major vessels (0-3) colored by fluoroscopy
- **thal**: Thalassemia (0-3. 1 = normal; 2 = fixed defect; 3 = reversable defect)
- **target**: Diagnosis of heart disease (1 = presence/high risk, 0 = absence/low risk)

## 5. Handle Missing Values and Duplicates
If there are missing values, we could impute them. Duplicates should be dropped.


```python
if df.duplicated().sum() > 0:
    df.drop_duplicates(inplace=True)
    print('Duplicates dropped. New shape:', df.shape)
else:
    print('No duplicates found.')
```

## 6. Outlier Detection and Treatment
We'll cap the extreme outliers for numeric continuous variables like `trestbps`, `chol`, `thalach` using the IQR method.


```python
def treat_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    return df

continuous_cols = ['trestbps', 'chol', 'thalach', 'oldpeak']
for col in continuous_cols:
    df = treat_outliers(df, col)
print('Outliers treated using IQR capping.')
```
